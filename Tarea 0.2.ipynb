# 3.-Crea una funcion que cree una grafica y dibuje 'k' lineas de forma $ f(x)=sin(\alpha x) $

    * x1 valor inicial de x    * 
x: valor final de''    * n: cantidad  e puntos del vecto x    * '
a: valor inicial \alphap    * n
b: valor final\alpha.    * pg
k: cantidad de líneas

```python
def foo(x1, x2, n, a, b, k):
    x = np.linspace(x1, x2, n)
    alfa = np.linspace(a, b, k)

    for alfa in alfa:
        y = np.sin(alfa * x)
        plt.plot(x, y)


plt.grid()
plt.xlabel("Eje x")
plt.ylabel("Eje y")

foo(0,2*np.pi,1000,1,3,3)
```
